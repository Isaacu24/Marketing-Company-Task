```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import random
import string
sns.set()
```

### .Load Dataset

```python
import pandas as pd
Data=pd.read_csv('marketing_company_data_task.csv')
Data
```

### .Inspect Dataset

```python
Data.info()
```

### Check for missing values

```python
Data.isna().sum()
```

### .Handle missing values by filling it up with mean, median or clearing them off

```python
Data= Data.fillna(Data['Clicks'].median())
Data
```

```python
Data= Data.fillna(Data['Impressions'].mean())
Data
```

### Format Data Type

```python
Data['Date']= pd.to_datetime(Data['Date'])
Data
```

## Task 1.

```python
sns.barplot(data=Data, x='Region', y='Conversions',)
plt.title('Conversions vs Region')
plt.xlabel('Region')
plt.ylabel('Conversions')
plt.show
```

## Task 2.

```python
sns.scatterplot(data=Data, x='Clicks', y='Impressions', palette='coolwarm', hue='Clicks')
plt.figure(figsize=(10,5))
plt.title('Clicks vs Impressions')
plt.xlabel('Clicks')
plt.ylabel('Impressions')
plt.show()
```

## Task 3.

```python
Region_count = Data ['Region'].value_counts()
plt.pie(Region_count, labels= Region_count.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Total Budget Allocation')
plt.show()
```

## Task 4.

```python
sns.lineplot(data=Data, x='Date',y='Impressions', palette='coolwarm',)
plt.figure(figsize=(10,5))
plt.xlabel('Impressions')
plt.ylabel('Date')
plt.title('Total Number')
plt.show
```

## Task 5.

```python
sns.histplot(Data['Budget (USD)'], bins=10, kde=True, color='red')
plt.title('Budget Distribution')
plt.xlabel('Budget (USD)')
plt.ylabel('Frequency')
plt.show
```

## Task 6.

```python
sns.scatterplot(data=Data, x='Budget (USD)', y='Conversions', palette='coolwarm', hue='Budget (USD)')
plt.figure(figsize=(10,5))
plt.title('Budget (USD) vs Conversions')
plt.xlabel('Budget (USD)')
plt.ylabel('Conversions')
plt.show()
```

## Task 8.

```python
sns.boxplot(x='Impressions', y='Region', data=Data, palette='Set3', hue='Region')
plt.show()
```

```python
sns.heatmap(Data[['Budget (USD)', 'Clicks', 'Impressions', 'Conversions']].corr(), annot=True, cmap='tab10')
plt.title('Correlation heatmap')
plt.show()
```
